```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
```

```python
# reading dataset (csv file)
data = pd.read_csv(r"C:\Users\Lenovo\Documents\jupyter_notebook\Churn_Modelling.csv",encoding='latin-1')
```

```python
df = pd.DataFrame(data) # dataframe is primary data structure of pandas
df
```

```python
# for churn prediction id,name,row num are not required (for training data)
df = df.drop(columns = ['RowNumber','CustomerId']) 
df
```

```python
# converting the genders male and female to 1 and 0(for good understanding of machine)
df['Gender'] = df.Gender.map({'Female':0,'Male':1})
df
```

```python
df.columns
```

```python
# checking null value present or not
df.isna().sum()
```

```python
# I used get_dummies to convert my text value of column(Geography) to numerical
df = pd.get_dummies(columns=['Geography'],data=df)
df.head()
```

```python
# when i used get_dummies i get true false as my values so to convert into numerical i used map method(0 and 1) 
df['Geography_France'] = df.Geography_France.map({ True : 1,False : 0})
df['Geography_Germany'] = df.Geography_Germany.map({ True : 1,False : 0})
df['Geography_Spain'] = df.Geography_Spain.map({ True : 1,False : 0})
df
```

```python
# for churn prediction id,name,row num are not required (for training data)
df = df.drop(columns = ['Surname'])
df
```

```python
df.columns
```

```python
# splitting my data for train and test
x = df[['CreditScore', 'Gender', 'Age', 'Tenure', 'Balance', 'NumOfProducts',
       'HasCrCard', 'IsActiveMember', 'EstimatedSalary',
       'Geography_France', 'Geography_Germany', 'Geography_Spain']]
y = df[['Exited']]
```

```python
# x has input values and y has the output values churn(here exited column)
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
# used random forest algorithm from supervised learning for classification
model = RandomForestClassifier()
```

```python
# training
model.fit(x_train,y_train)
```

```python
prediction = model.predict(x_test)
```

```python
from sklearn.metrics import classification_report
```

```python
# to check my accuracy and performance
print(classification_report(y_test,prediction))
```
